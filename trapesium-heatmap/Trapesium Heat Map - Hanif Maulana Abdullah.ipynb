### Membuat Heat Map (bebas)

dalam hal ini saya membuat Trapesium Heat Map (Level 2)

```python
#Trapesium Heat Map

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#Array (matriks) heat berukuran 9x9 dibuat secara manual, entri2nya diisi berdasarkan Discrete Mean-Value Property
#yang diselesaikan dengan sistem persamaan linear 14 variabel
#Bisa juga dicari dengan metode Forward in Time, Central in Space (FTCS)
#atau metode Crank-Nicolson) dalam persamaan diferensial parsial untuk menghitung solusi persamaan panas

#Array heat menyatakan suhu pada permukaan benda berbentuk trapesium
#Entri matriks yang berisi angka '6' bertujuan untuk membuat warna yang kontras dari bentuk trapesiumnya,
#dalam hal ini direpresentasikan dengan warna kuning

heat = np.array([[4, 6, 6, 6, 6, 6, 6, 6, 6],
                    [2, 4, 6, 6, 6, 6, 6, 6, 6],
                    [2, 3.19, 4, 6, 6, 6, 6, 6, 6],
                    [2, 2.77, 3.42, 4, 6, 6, 6, 6, 6],
                    [2, 2.45, 2.91, 3.4, 4, 6, 6, 6, 6],
                    [2, 2.14, 2.37, 2.71, 3.18, 4, 6, 6, 6],
                    [2, 1.72, 1.74, 1.86, 2.01, 2, 6, 6, 6],
                    [1, 1, 1, 1, 1, 1, 6, 6, 6],
                    [6, 6, 6, 6, 6, 6, 6, 6, 6]])
#suhu dalam derajat celcius dalam skala 1:6 suhu aslinya, dibulatkan dua angka di belakang koma

#untuk menampilkan heat map
fig, ax = plt.subplots()
im = ax.imshow(heat)

#Loop digunakan untuk membentuk tampilan angka pada bentuk segitiga sama sisi dengan panjang sisi yang sama 5 pada heat map
for i in range(5):
    for j in range((i+1)):
        text = ax.text(j, i, heat[i, j], ha="left", va="center", color="w")
        #color w=white, ha=horizontal axis, va=vertical axis

#Loop digunakan untuk membentuk tampilan angka pada baris ke-6 pada heat map
for i in range(1):
    for j in range(6):
        text = ax.text(j, i+5, heat[i+5, j], ha="left", va="center", color="w")
        
#Loop digunakan untuk membentuk tampilan angka pada baris ke-7 pada heat map
for i in range(1):
    for j in range(6):
        text = ax.text(j, i+6, heat[i+6, j], ha="left", va="center", color="w")

#Loop digunakan untuk membentuk tampilan angka pada baris ke-8 pada heat map
for i in range(1):
    for j in range(6):
        text = ax.text(j, i+7, heat[i+7, j], ha="left", va="center", color="w")

#Sehingga apabila kedua loop dijalankan, diperoleh tampilan angka berbentuk trapesium
#sesuai dengan heat map yang ditampilkan berdasarkan array yang dibuat
        
ax.set_title("Trapesium Heat Map")
fig.tight_layout()
plt.show()
```
