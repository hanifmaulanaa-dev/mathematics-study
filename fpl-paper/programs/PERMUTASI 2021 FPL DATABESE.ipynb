```python
print(KIPERAR)

#KIPERARRAT=array(zeros((2,50)))
KIPERARRAT=KIPERAR
#RATAAN POIN = TOTAL POIN MUSIM LALU / 38 (RATAAN PER PEMAIN)
for i in range (49): #ADA 50 KIPER
    KIPERARRAT[i][2]=KIPERAR[i][2]/38

print(KIPERARRAT)
    
#KIPERARVAL=array(zeros((2,50)))
KIPERARVAL=KIPERARRAT
#VALUE = PELUANG POIN * RATAAN POIN
#BIKIN ARRAY BARU, ISINYA NAMA, TIM, COST, RATAAN POIN (CUMA NINDIHIN POINTS DI ARRAY LAMA)
for i in range (49): #ADA 50 KIPER   
    if KIPERARVAL[i][1]=='Arsenal' and 'Chelsea':
        KIPERARVAL[i][2]=round(KIPERARRAT[i][2]*0.9,4)
    elif KIPERARVAL[i][1]=='Everton' and 'Liverpool' and 'Man City' and 'Newcastle' and 'Southampton' and 'Wolves':
        KIPERARVAL[i][2]=round(KIPERARRAT[i][2]*0.7,4)
    elif KIPERARVAL[i][1]=='Leicester' and 'Man Utd' and 'Spurs' and 'West Ham':
        KIPERARVAL[i][2]=round(KIPERARRAT[i][2]*0.5,4)
    elif KIPERARVAL[i][1]=='Aston Villa' and 'Brighton' and 'Burnley' and 'Fulham' and 'Leeds' and 'West Brom':
        KIPERARVAL[i][2]=round(KIPERARRAT[i][2]*0.3,4)
    else:
        KIPERARVAL[i][2]=round(KIPERARRAT[i][2]*0.1,4)
        
print(KIPERARVAL)

#URUTIN PEMAIN YG PUNYA VALUE TERTINGGI DI SETIAP POSISI
#DICARI VALUE MAX, DENGAN COST MIN
#COST <= 100

#ASCENDING SORT
sorted_array=KIPERARVAL[argsort(KIPERARVAL[:, 2])]
print('Sorted Array : ')
print(sorted_array)
```
