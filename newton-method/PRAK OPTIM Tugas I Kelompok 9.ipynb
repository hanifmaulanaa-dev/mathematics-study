## 0. Program Barisan Fibonacci
Berikut merupakan program mencari barisan fibonacci untuk n suku berhingga

```python
def Fibonacci(i): 
    #if n<0:
    #    return "Inputkan bilangan bulat positif"
    if i==0: 
        return 1 
    elif i==1: 
        return 1
    else: 
        return Fibonacci(i-1)+Fibonacci(i-2)

n=int(input("Masukkan n suku yang anda butuhkan : "))
while n<=0:
    n=int(input("Masukkan bilangan bulat positif! : "))

print(" ")
print("Diperoleh Barisan Fibonacci Berhingga",n,"Suku sebagai berikut:")

for i in range(n):
    print(Fibonacci(i))
```

## Interpretasi
Pertama-tama diberikan pilihan untuk memberikan *input* bilangan " n " yang menyatakan n suku yang dibutuhkan. Program akan terus mengulang untuk meminta masukan bilangan bulat positif apabila bilangan yang dimasukkan bukan bilangan bulat positif. **Sebagai catatan program hanya dapat menerima *input* berupa bilangan bulat, sebab telah didefinisikan dalam program, *input* harus bernilai *"int"* atau integer.** Dengan demikian, program akan langsung berhenti apabila *input* yang diberikan bukan berupa bilangan bulat.

Program di atas merupakan program untuk menentukan barisan fibonacci untuk suku n berhingga. Terlebih dahulu didefinisikan *function* Fibonacci terhadap n dengan pengeluaran berupa suatu bilangan, yaitu:
1. Bilangan 1 apabila i bernilai 0 **(i==0).** **Menyatakan suku pertama barisan Fibonacci yaitu 1**
2. Bilangan 1 apabila i bernilai 1 **(i==1).** **Menyatakan suku kedua barisan Fibonacci yaitu 2**
3. Jumlahan dua suku sebelumnya apabila i bernilai selain 0 dan 1. **Menyatakan suku ketiga, dan seterusnya dari barisan Fibonacci**

Setelah diberikan banyaknya n suku barisan yang dibutuhkan melalui *input*, program akan berjalan kemudian *output* akan diberikan melalui *loop* for yang bergerak dari i=0 sampai i=n. Hasil program ini berupa barisan fibonacci untuk suku n berhingga dalam bentuk menurun.

**Kelebihan :**
- Program ini dapat melakukan *input* secara berulang tanpa harus menjalankan ulang program apabila bilangan yang dimasukkan tidak sesuai dengan kriteria.

**Kekurangan :**
- Program ini hanya dapat digunakan untuk mencari barisan fibonacci dengan n suku berhingga.

# 1b. Program Peminimum Fungsi Menggunakan Metode Selang Tiga Titik
b. Berikut merupakan program mencari peminimum fungsi g dengan g(x)=x^4-4x^3+60x^2-70 menggunakan metode selang tiga titik

```python
from numpy import linspace
from pandas import DataFrame

def g(x):
    return(x**4 -4*x**3 +60*x**2 -70)

k=0
a=[0]
b=[2]
L=b[0]-a[0]
alpha=[]
beta=[]
gamma=[]
galpha=[]
gbeta=[]
ggamma=[]

while L>10**(-5):
    s=linspace(a[k],b[k],5)
    alpha.append(s[1])
    beta.append(s[2])
    gamma.append(s[3])
    galpha.append(g(alpha[k]))
    gbeta.append(g(beta[k]))
    ggamma.append(g(gamma[k]))
    
    if galpha[k]<gbeta[k]<ggamma[k]:
        a.append(a[k])
        b.append(beta[k])
    elif galpha[k]>gbeta[k]>ggamma[k]:
        a.append(beta[k])
        b.append(b[k])
    else:
        a.append(alpha[k])
        b.append(gamma[k])
    
    k=k+1
    L=(b[k]-a[k])/4

alpha.append(float('nan'))
beta.append(float('nan'))
gamma.append(float('nan'))
galpha.append(float('nan'))
gbeta.append(float('nan'))
ggamma.append(float('nan'))

xp=(b[k]+a[k])/2
gxp=g(xp)

print('Titik optimum f adalah (%f,%f)' %(xp,gxp))
data={'a':a,'alpha':alpha,'beta':beta,'gamma':gamma,'b':b,'g(alpha)':galpha,'g(beta)':gbeta,'g(gamma)':ggamma}
DataFrame(data)
```

#####   
Berikut merupakan grafik fungsi g dengan g(x)=x^4-4x^3+60x^2-70 pada interval x dari 0 sampai 2 dan x dari 0 sampai 0.5

```python
import matplotlib.pyplot as plt
from numpy import linspace

plt.figure(figsize=(15,4))
plt.suptitle("Grafik g(x)=x^4-4x^3+60x^2-70")

plt.subplot(121)
x=linspace(0,2,100)
x0=linspace(0,0.5,100)
xb=linspace(0.000015,0.000015,1)
plt.plot(x,g(x),label="g(x), x[0,2]",color='b')
plt.plot(x0,g(x0),label="g(x), x[0,0.5]",color='r')
plt.plot(xb,g(xb),'bo',color='black')
plt.title("Grafik I : g(x), x[0,2]")
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.1,2.1,-110,170])
plt.legend(loc='best')
plt.annotate('g(x) minumum', xy=(0,-88), xytext=(0,-88))
plt.grid()

plt.subplot(122)
x1=linspace(0,0.5,100)
x2=linspace(0.000015,0.000015,1)
plt.plot(x1,g(x1),label="zoom in",color='r')
plt.plot(x2,g(x2),'bo',color='black')
plt.title("Grafik II : g(x), x[0,0.5]")
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([-0.03,0.53,-73,-55])
plt.legend(loc='best')
plt.annotate('(%f,%f)' %(xp,gxp), xy=(0,-71.25), xytext=(0,-71.25))
plt.grid()

plt.show()
```

# Interpretasi
Program di atas merupakan program untuk menentukan peminimum fungsi g dengan g(x)=x^4-4x^3+60x^2-70 menggunakan metode selang tiga titik, serta program untuk membuat grafik fungsi g pada interval x[0,2] dan x[0,0.05].

Diambil ***linspace*** dari *package **numpy** dan ***DataFrame*** dari package* ***pandas***. *linspace* digunakan untuk membagi suatu interval menjadi beberapa bagian sesuai yang diinginkan, sedangkan *DataFrame* digunakan untuk membuat pengeluaran berupa tabel.

Terlebih dahulu didefinisikan :
1. bilangan k=0 sebagai nilai awal. k akan menjadi indeks yang bergerak dari k=0 sampai L<=10^(-5)
2. array a[] yang menyatakan batas bawah interval utama saat k. sebagai awalan dipilih a=[0]
3. array b[] yang menyatakan batas atas interval utama saat k. sebagai awalan dipilih b=[2]
4. L=b[0]-a[0] menyatakan panjang interval utama saat k=0.
5. array alpha[] yang menyatakan pemilihan titik pertama, namakan x1 dalam interval utama saat k.
6. array beta[] yang menyatakan pemilihan titik kedua, namakan x2 dalam interval utama saat k.
7. array gamma[] yang menyatakan pemilihan titik ketiga, namakan x3 dalam interval utama saat k.
8. array galpha[] yang menyatakan nilai fungsi g(x1) dalam interval utama saat k.
9. array galpha[] yang menyatakan nilai fungsi g(x2) dalam interval utama saat k.
10. array galpha[] yang menyatakan nilai fungsi g(x3) dalam interval utama saat k.

Pemilihan tiga titik x1, x2, x3 dengan x1 < x2 < x3 menginisiasi metode yang dilakukan oleh program ini yaitu **"Metode Selang Tiga Titik"**.

Selanjutnya, untuk setiap panjang interval utama L>10^(-5) diperoleh:
1. Jika g(x1) < g(x2) < g(x3), maka a[k+1]=a[k] dan b[k+1]=beta[k], untuk setiap indeks k.
2. Jika g(x1) > g(x2) > g(x3), maka a[k+1]=beta[k] dan b[k+1]=b[k], untuk setiap indeks k.
3. Jika kondisi 1 dan 2 tidak dipenuhi, maka a[k+1]=alpha[k] dan b[k+1]=gamma[k], untuk setiap indeks k.

Hal inilah yang disebut algoritme untuk **"Metode Selang Tiga Titik"**.

Sebagai tambahan, "alpha.append(float('nan'))", dkk. dipilih agar pada iterasi terakhir, nilai dari x1, dkk. diganti oleh NaN (Not a Number). Tujuannya untuk mengisi kekosongan pada tabel baris terakhir agar program tetap berjalan normal.

**Kelebihan :**
- Program ini dapat digunakan untuk mencari nilai pendekatan, bukan solusi analitiknya.

**Kekurangan :**
- Karena program ini hanya mencari nilai pendekatan dari solusi analitiknya, maka dalam program ini harus dipilih terlebih dahulu eror/toleransi agar program ini dapat berjalan dan selesai. Dalam hal ini dipilih erornya yaitu 10^(-5).
- Untuk membuat grafik terlihat menarik, dibutuhkan ketekunan yang tinggi. Hal ini terlihat dalam pemilihan koordinat untuk menandai titik minimum fungsi g beserta koordinatnya.

# 1c. Program Peminimum Fungsi Menggunakan Metode Selang Tiga Titik
c. Berikut merupakan program mencari peminimum fungsi h dengan h(x)=0.5x^2-sin(x) menggunakan metode selang tiga titik

```python
import numpy as np
from numpy import linspace
from pandas import DataFrame

def h(x):
    return(0.5*x**2 -np.sin(x))

k=0
a=[-np.pi]
b=[np.pi]
L=b[0]-a[0]
alpha=[]
beta=[]
gamma=[]
halpha=[]
hbeta=[]
hgamma=[]

while L>10**(-5):
    s=linspace(a[k],b[k],5)
    alpha.append(s[1])
    beta.append(s[2])
    gamma.append(s[3])
    halpha.append(h(alpha[k]))
    hbeta.append(h(beta[k]))
    hgamma.append(h(gamma[k]))
    
    if halpha[k]<hbeta[k]<hgamma[k]:
        a.append(a[k])
        b.append(beta[k])
    elif halpha[k]>hbeta[k]>hgamma[k]:
        a.append(beta[k])
        b.append(b[k])
    else:
        a.append(alpha[k])
        b.append(gamma[k])
    
    k=k+1
    L=(b[k]-a[k])/4

alpha.append(float('nan'))
beta.append(float('nan'))
gamma.append(float('nan'))
halpha.append(float('nan'))
hbeta.append(float('nan'))
hgamma.append(float('nan'))

xp=(b[k]+a[k])/2
hxp=h(xp)

print('Titik optimum h adalah (%f,%f)' %(xp,hxp))
data={'a':a,'alpha':alpha,'beta':beta,'gamma':gamma,'b':b,'h(alpha)':halpha,'h(beta)':hbeta,'h(gamma)':hgamma}
DataFrame(data)
```
